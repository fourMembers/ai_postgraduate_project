```python
def random_flip(img):
    x_flip = np.random.choice(a=[False, True])
    y_flip = np.random.choice(a=[False, True])
    z_flip = np.random.choice(a=[False, True])
    if x_flip:
        img = img[::-1,:,:]
    if y_flip:
        img = img[:,::-1,:]
    if z_flip:
        img = img[:,:,::-1]
    return img
```
