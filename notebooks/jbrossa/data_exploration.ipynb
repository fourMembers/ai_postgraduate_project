```python
train_path = 'D:/DL_postgraduate/final_project/data/Task07_Pancreas/imagesTr/'
all_data = os.listdir(train_path)
rm=[]

for i in range(len(all_data)):
  if all_data[i][0]==".":
    rm.append(all_data[i])

for element in rm:
    all_data.remove(element)

all_data
```

```python
from nibabel.testing import data_path
from nilearn import plotting
import nibabel as nib
```

```python
x_size = []
y_size = []
z_size = []
for data in all_data:
    img = nib.load(train_path + data)
    x_size.append(int(img.shape[0]))
    y_size.append(int(img.shape[1]))
    z_size.append(int(img.shape[2]))
```

```python
plt.bar(range(len(x_size)),x_size)
plt.show()
```
